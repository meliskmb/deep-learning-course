```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, ConfusionMatrixDisplay, classification_report
)

import warnings
warnings.filterwarnings('ignore')
```

```python
data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target, name='target')  # 0: malignant, 1: benign

df = X.copy()
df['target'] = y
df['diagnosis'] = df['target'].map({0: 'Malignant (Kötü Huylu)', 1: 'Benign (İyi Huylu)'})

print(f'Veri seti boyutu: {X.shape[0]} örnek, {X.shape[1]} öznitelik')
print(f'Benign: {(y==1).sum()}')
print(f'Malignant: {(y==0).sum()}')
```

```python
X.head()
```

```python
# Eğitim / Test ayrımı (%80 / %20)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, stratify=y)

print(f'Eğitim seti: {X_train.shape[0]} örnek')
print(f'Test seti: {X_test.shape[0]} örnek')

# StandardScaler ile normalizasyon
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc  = scaler.transform(X_test)
```

```python
# Model tanımlama ve eğitim
model = RandomForestClassifier(
    n_estimators=100,
    max_depth=None,
    random_state=42,
    n_jobs=-1
)

model.fit(X_train_sc, y_train)
print(f'Toplam karar ağacı sayısı: {model.n_estimators}')
```

```python
# Test seti üzerinde tahmin
y_pred = model.predict(X_test_sc)

# Metrikler
acc  = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec  = recall_score(y_test, y_pred)
f1   = f1_score(y_test, y_pred)

print(f'Accuracy (Doğruluk): {acc:.4f}')
print(f'Precision (Kesinlik): {prec:.4f}')
print(f'Recall (Duyarlılık): {rec:.4f}')
print(f'F1-Score: {f1:.4f}')

print(classification_report(y_test, y_pred, target_names=data.target_names))
```
