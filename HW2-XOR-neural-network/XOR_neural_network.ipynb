```python
import numpy as np

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])
```

```python
# Tek kkatmanlı
np.random.seed(42)
W = np.random.randn(2, 1) * 0.1
b = np.zeros((1, 1))
lr = 0.5

def sigmoid(x):
    return 1 / (1 + np.exp(-np.clip(x, -500, 500))) # overflow hatası alınmasın diye sırnıladım

for _ in range(2000):
    out = sigmoid(X @ W + b)
    W -= lr * (X.T @ (out - y)) / len(X)
    b -= lr * np.mean(out - y)

preds = (sigmoid(X @ W + b) > 0.5).astype(int)
for i in range(4):
    print(f"{X[i]} => beklenen: {y[i][0]}, tahmin: {preds[i][0]}")
print(f"Doğruluk: {np.mean(preds == y)*100:.1f}%")
```

```python
# Çok katmanlı
np.random.seed(42)
W1 = np.random.randn(2, 4) * np.sqrt(2/2)
b1 = np.zeros((1, 4))
W2 = np.random.randn(4, 1) * np.sqrt(2/4)
b2 = np.zeros((1, 1))
lr = 0.5

def relu(x): return np.maximum(0, x)
def relu_d(x): return (x > 0).astype(float)

for _ in range(10000):
    z1 = X @ W1 + b1
    a1 = relu(z1)
    z2 = a1 @ W2 + b2
    a2 = sigmoid(z2)

    # türev, zincir kuralı
    dz2 = (a2 - y) * a2 * (1 - a2)
    W2 -= lr * (a1.T @ dz2) / len(X)
    b2 -= lr * np.mean(dz2)

    dz1 = (dz2 @ W2.T) * relu_d(z1)
    W1 -= lr * (X.T @ dz1) / len(X)
    b1 -= lr * np.mean(dz1, axis=0)

preds = (a2 > 0.5).astype(int)
for i in range(4):
    print(f"{X[i]} => beklenen: {y[i][0]}, tahmin: {preds[i][0]}")
print(f"Doğruluk: {np.mean(preds == y)*100:.1f}%")
```
